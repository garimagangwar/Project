## Q 5. Find the Total Revenue generated by the E-Commerce company over all orders placed. (6 marks)

#### Hint: Calculate revenue of E-commerce company using the conditions mentioned above and then do the total (sum) of all to get total Revenue.

### Ans 5

First,we can create a function which takes the discounted_price as input and perform the logic **using the if else statement** and **return the total_revenue** at the end.
After adding the new column we can check for the **new column using info()** .

```python
def total_revenue  (discounted_price):
    if discounted_price > 600:
       total_revenue = discounted_price * 0.25
    elif discounted_price > 350:
         total_revenue = discounted_price * 0.15
    elif discounted_price > 100:
         total_revenue = discounted_price * 0.10
    else:
          total_revenue = discounted_price * 0.05
    return round (total_revenue,2)

ECom_Data['total_revenue']=ECom_Data['discounted_price'].apply(total_revenue)
ECom_Data.info()
```

Using the sum () over the newly created column we can calculate the total revenue.

```python
print("Total Brand Revenue is{}".format(ECom_Data['total_revenue'].sum()))
```

## Q 8. Create a new dataframe to include the Brand specific information as stated:
## i.	total number of orders placed per Brand
## ii.	total retail_price per Brand
## iii.	total discounted_price per Brand, and
## iv.	total BrandRevenue generated per Brand.

## Also, draw a pairplot using these four features. (6 marks)


### Ans 8

Here assuming we have DataFrame called **original_ECom_Data** that contains information about **orders,including columns**.so we can **create new DataFram**.
after that we calculate required satatics **using groupby()** method.
Now we can draw a **pairplot using seaborn** with the four features we have calculated.

```python
#sample dataset
data = {
    "Brand": ["BrandA","BrandB","BrandA","BrandB","BrandC"],
    "retail_price":[50,60,45,55,70],
    "discounted_price":[40,50,35,45,60],
    "order_quantity":[100,150,80,120,90]
}
original_ECom_Data = pd.DataFrame(data)
#Calculate the required statistics per Brand
brand_ECom_Data = original_ECom_Data.groupby("Brand").agg({
    "order_quantity":"sum",
    "retail_price":"sum",
    "discounted_price":"sum"
}).reset_index()
#calculate BrandRevenue per Brand
brand_ECom_Data['BrandRevenue'] = brand_ECom_Data["discounted_price"] * brand_ECom_Data["order_quantity"]
print(brand_ECom_Data)
```

Here,we draw a pair plot using four features **order_quantity,retail_price,discounted_price and brand_revenue**.

```python
#draw a pairplot
sns.pairplot(brand_ECom_Data, vars=["order_quantity","retail_price","discounted_price","BrandRevenue"],hue="Brand")
plt.show()
```

## Q 9.Compare performance regionwise

## a)	Draw a lineplot for the monthly Revenue of E-Commerce Company for each region separately.  (4 marks)

## b)	Identify the best and the worst performing months for each region. (2 marks)

#### Note: Only those days with actual orders(Order_Date) placed are present in the dataset. Assuming there were no orders on other days. Also, show Month and Year on x-axis.

### Ans 9 a)

Firstly Create a DataFrame and **group the data** by Region and date using **sum()** function then **plot the lineplot for each region** .Here we use plt.title for display the title of plot.

```python
# Create a DataFrame
data ={
     'Order Date':['2023-01-01','2023-01-02','2023-02-01','2023-02-02','2023-03-01'],
    'Region': ['RegionA','RegionB','RegionA','RegionB','RegionA'],
    'Revenue': [1000, 1500, 800, 1200, 2000]
}
monthly_revenue =ECom_Data.groupby(['Region', 'Order_Date']).sum().reset_index()
#plot the lineplot for each region
monthly_revenue.plot(kind='line', marker='o')
plt.title('Monthly Revenue by Region')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.legend(title='Region')
plt.show()
```

### Ans 9 b)

Here we find best and worst performing months using **Group the data by Region and order date.**

```python
#Group the data by Region and Order Date and find the best and worst performing months
best_months = ECom_Data.groupby(['Region',ECom_Data['Order_Date']]).sum()
worst_months = ECom_Data.groupby(['Region',ECom_Data['Order_Date']]).sum()

print("Best Performing Months:")
print(best_months)

print("\nWorst performing Months:")
print(worst_months)
```
